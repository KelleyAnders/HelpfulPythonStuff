```python
#Read in the file
in_df = pd.read_csv('file.csv', encoding='utf-8', low_memory=False)
#Set the column you want to create the word cloud rom to string type
in_df['column'] = in_df['column'].astype(str)
in_df.info()
```

```python
# Generate the wordcloud
from wordcloud import STOPWORDS
stop_words = ["add more here"] + list(STOPWORDS)
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
wordcloud2 = WordCloud(stopwords = stop_words, background_color="white", width=1600, height=800).generate(' '.join(in_df['column']))
plt.figure(figsize=(20, 10))
plt.imshow(wordcloud2, interpolation='bilinear')
plt.axis("off")
plt.show()
```
